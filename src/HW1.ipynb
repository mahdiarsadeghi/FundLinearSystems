```julia
using Plots
```

# 1

```julia
function x(n)
    N = 6
    if n >= 0 
        if n <= N
            x = 0.8^n
            return x
        end
    end
    return 0
end

n = -10:10
scatter(n, x.(n), xlabel="n", label="x[n]", title="Question 1")
scatter!(n, x.(n .-2), label="x[n-2]")
scatter!(n, x.(-2*n .+3), label="x[-2n+3]")
```

# 2

```julia
function x(n)
    N = 4
    2*sin(*n/N) + cos(3**n/N)
end

function x(n)
    N = 4
    2*sin(3*n/N) + cos(9*n/N)
end

n = 0:16
scatter(n, x.(n), xlabel="n", label="x[n]")
plot!(n, x.(n), label=:false)
scatter!(n, x.(n), xlabel="n", label="x[n]")
plot!(n, x.(n), label=:false, title="Question 2")
```

# 3

```julia
function a(t)
    a, T,  = [0, 10, 0]
    exp(a*t)*cos(2**t/T+)
end

t = 0:0.01:50
p = plot(t,a.(t), xlabel="t", ylabel="x(t)", label="a", linewidth=2)

function b(t)
    a, T,  = [0, 10, -/4]
    exp(a*t+2*im**t/T+im*)
end
p = plot!(p, t,real.(b.(t)), xlabel="t", label="b", linewidth=2)

function c(t)
    a, T,  = [-0.05, 10, -/4]
    exp(a*t+2*im**t/T+im*)
end
p = plot!(t,real.(c.(t)), xlabel="t", label="c", linewidth=2, title="Question 3")
```

```julia
N   = 3
x(n) = cos(2**n/N)
n = 0:2*N
scatter(n, x.(n), label="x[n]", xlabel="n", title="Question 4")
```

```julia
x(n,k) = cos(2**k*n/N)
k = 1; p1 = scatter(n, x.(n,k), label="x"*string(k)*"[n]", xlabel="n")
k = 2; p2 = scatter(n, x.(n,k), label="x"*string(k)*"[n]", xlabel="n")
k = 3; p3 = scatter(n, x.(n,k), label="x"*string(k)*"[n]", xlabel="n")
k = 4; p4 = scatter(n, x.(n,k), label="x"*string(k)*"[n]", xlabel="n")
plot(p1,p2,p3,p4, layout=(2,2))
```
